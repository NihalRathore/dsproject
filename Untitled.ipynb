```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
iris_df = pd.DataFrame(iris)
# Display summary statistics of the dataset
iris_df.describe()

# Create a scatterplot matrix of the dataset using seaborn
sns.pairplot(iris_df, hue='species')
plt.show()
# Create a histogram of sepal length for each species
sns.histplot(data=iris_df, x='sepal_length', hue='species', kde=True)
plt.show()

# Create a kernel density plot of petal width for each species
sns.kdeplot(data=iris_df, x='petal_width', hue='species')
plt.show()
# Create a scatterplot of sepal length vs petal length
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species')
plt.show()

# Create a regression plot of sepal width vs petal length
sns.regplot(data=iris_df, x='sepal_width', y='petal_length', color='g')
plt.show()
```
